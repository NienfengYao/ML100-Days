## Numpy 亂數生成資料
* [numpy.random.rand](https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.random.rand.html#numpy.random.rand) (均勻分佈)
  * Create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1)
* [numpy.random.normal](https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.random.normal.html) (常態分佈)
  * Draw random samples from a normal (Gaussian) distribution.

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
"""
Your Code Here
"""
matrix = 2 * np.random.rand(10, 10) - 1
print(matrix[0, :])
plt.figure(figsize=(10,10))
"""
Your Code Here
"""
sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, annot = True)

plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = 2 * np.random.rand(nrow, ncol) - 1
# print(matrix)
# print(matrix.shape)

indice = np.random.choice([0,1,2], size=nrow)
# print(indice)
# print(indice.shape)
plot_data = pd.DataFrame(matrix, indice)
# print(plot_data)
# print(plot_data.shape)

grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(plt.hist)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.normal(loc=0.0, scale=1.0, size=(nrow, ncol))  # Draw random samples from a normal (Gaussian) distribution.

indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)


grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(plt.hist)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
